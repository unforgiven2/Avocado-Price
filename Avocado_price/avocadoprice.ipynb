```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("avocado.csv")
df.drop("Unnamed: 0",axis=1, inplace=True)
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df
```

```python
df["AveragePrice"].mean()
```

```python
list(set(df["region"].values.tolist()))
```

```python
df["region"].unique()
```

```python
l=df.groupby("region")["AveragePrice"].mean() 
l
```

```python
print("max price {} ".format(l.max()))
print("min price {}".format(l.min()))
print("avg price {}".format(l.mean()))
```

```python
fig, ax =plt.subplots(figsize=(30,15))

ax.scatter( x=df["AveragePrice"], y=df["Date"])
```

```python
c=df.groupby("type")["AveragePrice"].mean() 
c
```

```python
print("Average avocado price per year {}".format(df.groupby("year")["AveragePrice"].mean()))
print(("Minimum avocado price per year {}".format(df.groupby("year")["AveragePrice"].min())))
print(("Maximum avocado price per year {}".format(df.groupby("year")["AveragePrice"].max())))
```

```python
mask = df['type']=='conventional'
g = sns.factorplot('AveragePrice','region',data=df[mask],
                   hue='year',
                   size=8,
                   aspect=0.6,
                   join=False,
              )
```
